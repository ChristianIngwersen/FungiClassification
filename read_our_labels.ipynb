# Read our labels

```python
import os
from glob import glob

import pandas as pd
import numpy as np
```

```python
folder_path = '../sort_fungi_v1/'
classes_to_read = [0]
```

```python
img_paths = pd.DataFrame(glob(folder_path + '/**/*.JPG', recursive=True), columns=['image'])
img_paths['image'] = img_paths['image'].str.replace('\\', '/', regex=False)
img_paths['class'] =  img_paths['image'].str.split('/').str[-2]
img_paths['class'] = img_paths['class'].astype(int)
img_paths = img_paths[img_paths['class'].isin(classes_to_read)]
img_paths = img_paths[~img_paths['image'].str.contains('/0_')]
img_paths['image'] =  img_paths['image'].str.split('/').str[-1]
img_paths
```

```python
our_label_path = 'data_with_our_labels.csv'

if os.path.exists(our_label_path):
    df_labels = pd.concat((pd.read_csv(our_label_path), img_paths))
    df_labels = df_labels.drop_duplicates()
else:
    df_labels = img_paths
df_labels.to_csv(our_label_path, index=False)
```
