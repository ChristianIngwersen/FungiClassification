# Read our labels

```python
import os
from glob import glob

import pandas as pd
import numpy as np
```

```python
folder_path = '../sort_fungi_v1/'
classes_to_read = [0]
```

```python
img_paths = pd.DataFrame(glob(folder_path + '/**/*.JPG', recursive=True), columns=['image'])
img_paths['image'] = img_paths['image'].str.replace('\\', '/', regex=False)
img_paths['class'] =  img_paths['image'].str.split('/').str[-2]
img_paths['class'] = img_paths['class'].astype(int)
img_paths = img_paths[img_paths['class'].isin(classes_to_read)]
img_paths = img_paths[~img_paths['image'].str.contains('/0_')]
img_paths
```
